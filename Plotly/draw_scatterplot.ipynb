***@Title : Draw scatterplot for 1000 random number by plotly***

```python
import plotly
import plotly.graph_objs as go
import numpy as np
from plotly.offline import iplot
```

```python
n = 1000
random_x = np.random.randn(n)
random_y = np.random.randn(n)
```

```python
trace = go.Scatter(
    x = random_x,
    y = random_y,
    mode = 'markers'
)
data = [trace]
figure = go.Figure(data=data)
iplot(figure)
```
