***@Title : Draw scatterplot and add styles***

```python
import plotly
import plotly.graph_objs as go
import numpy as np
from plotly.offline import iplot
```

```python
n = 500
random_x = np.random.randn(n)
random_y = np.random.randn(n)
```

```python
trace = go.Scatter(
    x = random_x,
    y = random_y,
    mode = 'markers',
    marker =dict(color = 'Blue',size=10,line =dict(color='Red',width=1.5))
)
data = [trace]
figure = go.Figure(data=data)
iplot(figure)
```
