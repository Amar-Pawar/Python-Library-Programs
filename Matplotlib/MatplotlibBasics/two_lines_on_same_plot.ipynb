***@Title : Draw two lines on same plot and give legend***

```python
from matplotlib import pyplot as plt
```

```python
age_x = [23,24,25,26,27,28,29,30]
```

```python
py_devop = [2000,3000,3233,3500,3790,4500, 6789, 8900]
jv_devop = [1500,2200,2500,2700,3200,3800, 4700, 7000]
```

```python
plt.plot(age_x,py_devop, color='b', label='Python')
plt.plot(age_x,jv_devop,color='g',label='Java')
plt.title("Salary cap in USD")
plt.xlabel("Ages")
plt.ylabel("salary")
plt.legend()
plt.show()
```
