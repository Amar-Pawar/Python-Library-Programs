***@Title : Draw line chart and dsiplay grid***

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('financial_data.csv', index_col='Date')
```

```python
plt.plot(df)
plt.grid(color = 'black', linestyle = '--')
plt.xlabel('Dates')
plt.ylabel('Financial Data')
plt.title('Financial Data Analysis')
plt.legend(df.columns)
plt.show()
```
