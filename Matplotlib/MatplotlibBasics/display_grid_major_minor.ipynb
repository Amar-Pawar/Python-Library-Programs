***@Title : Draw line chart and dsiplay grid with major and minor lines***

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('financial_data.csv', index_col='Date')
```

```python
plt.plot(df)
plt.minorticks_on()
plt.grid(which='major', linestyle='--', linewidth='0.5', color='red')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color = 'black')
plt.xlabel('Dates')
plt.ylabel('Financial Data')
plt.title('Financial Data Analysis')
plt.legend(df.columns)
plt.show()
```
