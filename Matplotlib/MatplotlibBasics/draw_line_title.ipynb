***@Title : Draw line with suitable x and y axis***

```python
from matplotlib import pyplot as plt
```

```python
x = [1,4,2,7,10]
y = [4,2,6,1,8]
```

```python
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Basic Plot')
plt.show()
```
