***@Title : Display bar chart of languages and popularity and give specific position and width to bar***

```python
from matplotlib import pyplot as plt
```

```python
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
```

```python
plt.style.use('seaborn-darkgrid')
plt.xlabel('Languages')
plt.ylabel('Popularity')
plt.title('Languages with popularity')
width = [0.4,0.2,0.7,1.1,0.2,1.5]
x_position = [1,3,5,7,9,11]
plt.xticks(x_position, languages)
plt.bar(x_position, popularity, width=width)
plt.show()
```
