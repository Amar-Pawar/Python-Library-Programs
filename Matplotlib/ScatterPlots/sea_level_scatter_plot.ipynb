***@Title : Draw scatter plot for sea level for past 100 years***

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('sea_levels.csv')
year = data['year']
sea_levels = data['CSIRO_sea_level']
```

```python
plt.scatter(year, sea_levels,color='Blue', edgecolors='Black')
plt.xlabel('Year')
plt.ylabel('Sea level (inches)')
plt.title('Rise in sea level since 1880')
plt.show()
```
