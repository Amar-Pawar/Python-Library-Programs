```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
wt_group1=[50,53.7,37.5,87.2,55.8]
ht_group1=[147,154,167,165,174] 
wt_group2=[55,66,87,100,46]
ht_group2=[155.5,134.6,156,170,155] 
wt_group3=[55,77,65,45,98]
ht_group3=[165,156,147.5,167.3,150]
```

```python
weight=np.concatenate((wt_group1,wt_group2,wt_group3))
height=np.concatenate((ht_group1,ht_group2,ht_group3))
colors = ['b'] * 5 + ['g'] * 5 + ['r']*5
plt.scatter(weight, height, marker='x', color = colors)
plt.xlabel('Weight', fontsize=16)
plt.ylabel('Height', fontsize=16)
plt.title('Group wise weight and height scatter plot')
plt.show()
```
