***@Title : Draw seaborn violin plot for given data for day against sex***

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
```

```python
sns.violinplot(data=df, x="sex", y="total_bill")
plt.title('Plot for sex against total bill')
plt.xlabel('Sex')
plt.ylabel('Days')
sns.despine()
plt.show()
```
