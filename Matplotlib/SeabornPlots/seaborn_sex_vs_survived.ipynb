***@Title : Draw seaborn plot for sex vs survived for given data***

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('titanic')
```

```python
sns.barplot(data=df, x="sex", y="survived")
plt.title('Sex and Survived of Titanic Passengers')
plt.xlabel('sex')
plt.ylabel('survived')
sns.despine()
plt.show()
```
