***@Title : Draw seaborn swarmplot from given data for bill against day***

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
```

```python
sns.swarmplot(data=df, y="total_bill", x="day", size=2.5)
plt.title('Days vs total bill')
plt.xlabel('Days')
plt.ylabel('Total bill')
sns.despine()
plt.show()
```
