***@Title : Draw seaborn scatter plot for given data for day against bill***

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
```

```python
sns.scatterplot(data=df, x="day", y="total_bill")
plt.title('Total bill for days')
plt.xlabel('Days')
plt.ylabel('Total bill')
sns.despine()
plt.show()
```
