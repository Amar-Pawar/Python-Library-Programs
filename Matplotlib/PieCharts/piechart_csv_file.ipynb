***@Title : Read data from csv file and create pie chart***

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df =  pd.read_csv('medal.csv')
country_data = df["country"]
medal_data = df["gold_medal"]
```

```python
colors =['Red', 'Blue', 'Green','Grey','Brown']
explode = (0.1, 0, 0, 0, 0, )
```

```python
plt.pie(medal_data, labels=country_data, explode=explode, colors=colors,
autopct='%1.1f%%', shadow=True)
plt.title("Gold medal percentage of countries in 2016 olympics")
plt.show()
```
